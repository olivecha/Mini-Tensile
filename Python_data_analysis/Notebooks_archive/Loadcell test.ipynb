```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path_10 = '/Users/Olivier/Desktop/force_sensibility.txt'
path_5 = '/Users/Olivier/Desktop/force_sens_5.txt'
df_10 = pd.read_csv(path_10, sep=' ', names = ['strain', 'force'])
df_5 = pd.read_csv(path_5, sep=' ', names = ['strain', 'force'])
```

```python
df = df_10
x = range(0,df.shape[0])
plt.figure(figsize=(12,8))
plt.plot(x, df['force'], color='r', label='Measured Value')
mean1 = df['force'][:15].mean()
std1 = df['force'][:15].std()
mean2 = df['force'][17:].mean()
std2 = df['force'][17:].std()

plt.plot([0, 15], [mean1, mean1],'k')
plt.plot([0, 14], [mean1+std1, mean1+std1], 'k--')
plt.plot([0, 14], [mean1-std1, mean1-std1], 'k--')

plt.plot([17, 43], [mean2, mean2], c='k', label = 'Average Value')
plt.plot([18, 42], [mean2 + std2, mean2 + std2], 'k--', label='Standard Deviation')
plt.plot([18, 42], [mean2 - std2, mean2 - std2], 'k--')

plt.ylabel('Measured force (N)')
plt.xlabel('Data point index')
plt.title('Mesured force with loadcell when appliying a -10g delta')
plt.legend()
plt.show()
```

```python
df = df_5
x = range(0,df.shape[0])
plt.figure(figsize=(12,8))
plt.plot(x, df['force'], color='r', label='Measured Value')

mean1 = df['force'][:20].mean()
std1 = df['force'][:20].std()

mean2 = df['force'][25:100].mean()
std2 = df['force'][25:100].std()

mean3 = df['force'][110:].mean()
std3 = df['force'][110:].std()

plt.plot([0, 20], [mean1, mean1],'k')
plt.plot([0, 19], [mean1+std1, mean1+std1], 'k--')
plt.plot([0, 19], [mean1-std1, mean1-std1], 'k--')

plt.plot([25, 100], [mean2, mean2], c='k', label = 'Average Value')
plt.plot([25, 98], [mean2 + std2, mean2 + std2], 'k--', label='Standard Deviation')
plt.plot([25, 98], [mean2 - std2, mean2 - std2], 'k--')

plt.plot([105, 160], [mean3, mean3], c='k')
plt.plot([107, 160], [mean3 + std3, mean3 + std3], 'k--')
plt.plot([107, 160], [mean3 - std3, mean3 - std3], 'k--')

plt.ylabel('Measured force (N)')
plt.xlabel('Data point index')
plt.title('Mesured force with loadcell when appliying a +/- 5g delta')
plt.legend()
plt.show()
```
