```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n_tests = 12
paths = ['/Users/Olivier/Desktop/Mini Tract/Tests caractérisation machine 1/Data/max_force_'+str(i)+'.txt' for i in range(1,n_tests+1)]
dfs = [pd.read_csv(path, sep=' ', names = ['strain', 'force']) for path in paths]
N_steps = [4, 8, 16, 32, 32, 32, 16, 16, 8, 8, 4, 4]
C = [2.8, 2.8, 2.8, 2.8, 3.0, 3.5, 3.5, 3.0, 3.5, 3.0, 3.5, 3.0]
max_F = [df['force'].max() for df in dfs]
```

```python
# Nsteps = 4
# C = 2.8 A
df = dfs[0]
plt.figure()
index = range(0, len(df['strain']))
max_force = df['force'].max()
plt.plot(index, df['force'], color='k')
plt.scatter(df['force'].idxmax(), df['force'].max(), c='r')
plt.annotate('F = ' + str(df['force'].max()) + 'N', (df['force'].idxmax()+1, df['force'].max()))
plt.xlabel('index (~time)')
plt.ylabel('Force (N)')
plt.title('Max force Test with 5A power supply, 4steps/rev, 2.8A at drive')
plt.grid('on')
plt.show()
```

```python
# Dataframe for the max values
max_df = pd.DataFrame(data=np.array([C, N_steps, max_F]).transpose(), columns=['C', 'steps', 'force'])
max_df = max_df.sort_values(['C', 'steps'])

X = np.array(max_df['C']).reshape(3,4)
Y = np.array(max_df['steps']).reshape(3,4)
Z = np.array(max_df['force']).reshape(3,4)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, color='w')

ax.set_xlabel('Current (A)')
ax.set_ylabel('Steps/rev')
ax.set_zlabel('Max force')

plt.show()
```
