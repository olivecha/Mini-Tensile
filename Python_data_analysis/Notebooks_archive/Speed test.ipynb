```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file1 = '/Users/Olivier/Desktop/speed_50.txt'
file2 = '/Users/Olivier/Desktop/speed_100.txt'

df50 = pd.read_csv(file1, sep= ' ', names=['pos', 'time'])
df100 = pd.read_csv(file2, sep= ' ', names=['pos', 'time'])
```

```python
# Test at 100 milis per step
time = (df100['time'] - df100['time'].min())/1000
y = df100['pos'] - df100['pos'].min()
speed = np.gradient(y, time)

# Position plot
plt.figure(figsize=(10,8))
plt.plot(time, y, c='k')
plt.title('Position vs Time at 100 microseconds per step')
plt.xlabel('time (s)')
plt.ylabel('position (mm)')
plt.show()

# Speed plot
plt.figure(figsize=(10,8))
avg_speed = speed[:100].mean()
plt.plot(time, speed, c='k', label='measured speed')
plt.plot([0,11], [avg_speed, avg_speed], c='r', label = 'average speed = ' + str(np.around(avg_speed,0)) + ' mm/s')
plt.title('Speed vs Time at 100 microseconds per step')
plt.xlabel('time (s)')
plt.ylabel('Speed (mm/s)')
plt.legend()
plt.show()
```

```python
# Test at 50 milis per step
time = (df50['time'] - df50['time'].min())/1000
y = df50['pos'] - df50['pos'].min()
speed = np.gradient(y, time)

# Position plot
plt.figure(figsize=(10,8))
plt.plot(time, y, c='k')
plt.title('Position vs Time at 50 microseconds per step')
plt.xlabel('time (s)')
plt.ylabel('position (mm)')
plt.show()

# Speed plot
plt.figure(figsize=(10,8))
avg_speed = speed.mean()
plt.plot(time, speed, c='k', label='measured speed')
plt.plot([0,5], [avg_speed, avg_speed], c='r', label = 'average speed = ' + str(np.around(avg_speed,0)) + ' mm/s')
plt.title('Speed vs Time at 50 microseconds per step')
plt.xlabel('time (s)')
plt.ylabel('Speed (mm/s)')
plt.legend()
plt.show()
```
