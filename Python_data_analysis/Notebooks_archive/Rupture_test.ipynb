```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal
```

```python
base = 32.25
L0 = base - 21.75
Y = np.array([20.75, 19, 18, 17, 16, 15, 14, 13, 12, 11, 10, 9, 8])
distance = base - Y
DL = distance - L0
STh = DL / L0
S = [0.09, 0.25, 0.36, 0.44, 0.54, 0.63, 0.71, 0.81, 0.91, 0.99, 1.09, 1.20, 1.28]
x = range(len(S))
```

```python
plt.figure(figsize=(12,8))
plt.plot(x, S, label='measured strain')
plt.plot(x, STh, label='theoretical strain')
plt.legend()
plt.xlabel('data index')
plt.ylabel('strain')
plt.grid('on')
plt.show()

```
