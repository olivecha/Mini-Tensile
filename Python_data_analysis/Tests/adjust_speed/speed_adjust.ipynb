```python
import numpy as np 
import pandas as pd
import scipy.signal
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('test2.txt', sep=' ', decimal='.', names=['D', 'T']).astype(float)
df['D'] = scipy.signal.savgol_filter(df['D'], df['D'].shape[0]//5 + 1, 1)
```

```python
plt.plot(df['T'], df['D'])
```

```python
plt.plot(np.gradient(df['D'], df['T']))
```

```python
current_sped = np.mean(np.gradient(df['D'], df['T'])[100:500])
current_sped_mic_s = current_sped * 1000 * 1000 #mm/milis -> micron/s
current_sped_mic_s
```
